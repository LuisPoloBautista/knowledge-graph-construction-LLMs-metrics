## DESCARGA DE ARTÍCULOS

```python
from gdeltdoc import Filters, near, repeat


# Fecha de inicio y fin
start_date = datetime.date(2019, 11, 6)
end_date = datetime.date(2024, 9, 5) #2017, 1, 1

gd = GdeltDoc()

while start_date <= end_date:

  # Crear un nuevo conjunto de filtros para cada iteración
  filters = Filters(
      keyword= ['earthquake', 'tsunami', 'richter scale', 'hurricane', 'epicenter', 'flood', 'tectonic plates'],
      #keyword=['earthquake', 'richter scale', 'epicenter', 'tectonic plates', 'earthquakes'],
      #keyword=['Hurricane Manuel', 'Hurricane Patricia', 'Hurricane Odile', 'Hurricane Enrique', 'Hurricane Nora', 'Hurricane Lane', 'Hurricane Harvey', 'Hurricane Celia' ],
      start_date=start_date.strftime("%Y-%m-%d"),
      end_date=(start_date + datetime.timedelta(days=1)).strftime("%Y-%m-%d"),
      country="MX",
      #repeat = repeat(2, 'Mexico'),
      theme='NATURAL_DISASTER_TREMORS',
      num_records=250,
      )
  # Buscar artículos que coincidan con los filtros
  articles = gd.article_search(filters)

  # Crear un DataFrame con los resultados
  df = pd.DataFrame(articles)

  # Guardar los datos en un archivo CSV
  filename = f"articles_{start_date}.csv"
  df.to_csv('/content/drive/MyDrive/GDELT_API3/' + filename, index=False)

  # Incrementar la fecha de inicio para la próxima búsqueda
  start_date += datetime.timedelta(days=1)

  # Esperar 5 segundos antes de la próxima consulta
  time.sleep(5)
```

# CONCATENAR

```python
import os
import pandas as pd

# Ruta de la carpeta que contiene los archivos CSV
carpeta_csv = "/content/drive/MyDrive/GDELT_API3/"

# Ruta del archivo resultante
archivo_resultante = "/content/drive/MyDrive/GDELT_API3/Datos/GDELT_DN.csv"

# Lista para almacenar los DataFrames de cada archivo CSV válido
dfs = []

# Procesar cada archivo CSV en la carpeta
for archivo_csv in os.listdir(carpeta_csv):
    # Verificar si el archivo termina en ".csv" y no está vacío
    if archivo_csv.endswith(".csv") and os.path.getsize(os.path.join(carpeta_csv, archivo_csv)) > 0:
        try:
            # Leer el archivo CSV completo
            df = pd.read_csv(os.path.join(carpeta_csv, archivo_csv))
            # Verificar si el DataFrame está vacío
            if not df.empty:
                dfs.append(df)
        except pd.errors.ParserError as e:
            print(f"Error al leer el archivo {archivo_csv}: {e}")
        except Exception as e:
            # Manejar otras excepciones genéricas
            print(f"Error al procesar el archivo {archivo_csv}: {e}")
        else:
            print(f"Archivo {archivo_csv} procesado correctamente.")

# Concatenar todos los DataFrames en uno solo si hay DataFrames válidos
if dfs:
    df_concatenado = pd.concat(dfs, ignore_index=True)

    # Guardar el resultado en un nuevo archivo CSV, especificando el delimitador
    df_concatenado.to_csv(archivo_resultante, index=False)

    print(f"Archivos concatenados guardados en: {archivo_resultante}")
else:
    print("No se encontraron archivos CSV válidos para procesar.")
```

```python
from newspaper import Article
import csv
import pandas as pd

# Lee el archivo CSV en chunks y procesa cada uno
csv_file_path = '/content/drive/MyDrive/GDELT_API3/Datos/GDELT_DN.csv'
output_folder_path = '/content/drive/MyDrive/GDELT_API3/Datos/texto_completo/'

def process_chunk(chunk, output_file_path):
    texts = []  # Lista para almacenar los textos completos de cada URL
    for index, row in chunk.iterrows():
        try:
            url = row['url']
            article = Article(url)
            article.download()
            article.parse()
            text = article.text.replace(',', ' ').replace('\n', ' ')
            texts.append(text)
        except Exception as e:
            texts.append('')  # Si hay un error, agregar cadena vacía

    # Agrega la columna 'TEXT' al DataFrame del chunk
    chunk['texto_completo'] = texts

    # Guarda el resultado parcial en un nuevo archivo CSV
    chunk.to_csv(output_file_path, mode='a', index=False)

# Especifica el tamaño de cada chunk
chunk_size = 10
chunk_number = 0

for chunk in pd.read_csv(csv_file_path, chunksize=chunk_size):
    chunk_output_path = f'{output_folder_path}resultado_chunk_{chunk_number}.csv'
    process_chunk(chunk, chunk_output_path)
    chunk_number += 1

    # Libera memoria RAM
    del chunk

print("Procesamiento completo")
```

#Similitud de coseno

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
from nltk.corpus import stopwords
import nltk
```

```python
nltk.download('stopwords')
stop_words_es = stopwords.words('spanish')
```

```python
def remove_duplicate_news(csv_file, output_file, similarity_threshold=0.9):
    # Cargar el archivo CSV
    df = pd.read_csv('/content/drive/MyDrive/GDELT_API3/Datos/GDELT_DN_texto_completo.csv')

    # Asegurarse de que la columna de texto se llama 'text'
    if 'texto_completo' not in df.columns:
        raise ValueError("La columna 'text' no se encuentra en el archivo CSV.")

    # Crear el vectorizador TF-IDF con parámetros que reducen el uso de memoria
    vectorizer = TfidfVectorizer(stop_words=stop_words_es, max_features=10000, max_df=0.95)

    # Convertir los textos a una matriz TF-IDF
    tfidf_matrix = vectorizer.fit_transform(df['texto_completo'])

    # Calcular la similitud de coseno
    cosine_similarities = cosine_similarity(tfidf_matrix)

    # Identificar duplicados y calcular el porcentaje
    num_articles = len(cosine_similarities)
    to_remove = set()

    for i in range(num_articles):
        for j in range(i + 1, num_articles):
            if cosine_similarities[i, j] > similarity_threshold:
                to_remove.add(j)

    num_duplicates = len(to_remove)
    percentage_duplicates = (num_duplicates / num_articles) * 100

    # Mostrar porcentaje de artículos similares
    print(f"De los {num_articles} artículos, {num_duplicates} tienen contenido similar ({percentage_duplicates:.2f}%).")
    user_input = input("¿Deseas eliminar estos artículos? (s/n): ").strip().lower()

    if user_input == 's':
        # Filtrar el DataFrame para eliminar duplicados
        df_filtered = df.drop(index=list(to_remove))

        # Guardar el DataFrame resultante en un nuevo archivo CSV
        df_filtered.to_csv(output_file, index=False)
        print(f"Archivo CSV sin duplicados guardado como '{output_file}'.")
    else:
        print("No se realizaron cambios en el archivo.")

# Usar la función
remove_duplicate_news('noticias.csv', '/content/drive/MyDrive/GDELT_API3/Datos/noticias_sin_duplicados.csv', similarity_threshold=0.9)
```

#Filtrar por México

```python
df_filtro_m = pd.read_csv('/content/drive/MyDrive/GDELT_API3/Datos/noticias_sin_duplicados.csv')
df_filtro_m.info()
```

```python
# Lista de estados de México
estados_mexico = [
    'Aguascalientes', 'Baja California', 'Baja California Sur', 'Campeche', 'Chiapas',
    'Chihuahua', 'Coahuila', 'Colima', 'Durango', 'Guanajuato', 'Guerrero', 'Hidalgo',
    'Jalisco', 'México', 'Michoacán', 'Morelos', 'Nayarit', 'Nuevo León', 'Oaxaca',
    'Puebla', 'Querétaro', 'Quintana Roo', 'San Luis Potosí', 'Sinaloa', 'Sonora',
    'Tabasco', 'Tamaulipas', 'Tlaxcala', 'Veracruz', 'Yucatán', 'Zacatecas', 'Ciudad de México'
]

# Términos relacionados con México
terminos_mexico = ['México', 'Mexicana', 'Mexicano', 'CDMX', 'EDOMEX']
```

```python
import unicodedata

# Columna de texto a analizar
columna_texto = 'texto_completo'  # Reemplaza con el nombre real de la columna

# Función para eliminar acentos
def eliminar_acentos(texto):
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )

# Función para filtrar registros
def filtrar_registros(texto):
    # Handle potential non-string values
    if isinstance(texto, str):
        # Convertir a minúsculas y eliminar acentos
        texto = eliminar_acentos(texto.lower())
        for estado in estados_mexico:
            # Comparar también sin acentos
            if eliminar_acentos(estado.lower()) in texto:
                return True
        for termino in terminos_mexico:
            if eliminar_acentos(termino.lower()) in texto:
                return True
    return False

# Aplicar filtro
df_filtrado = df_filtro_m[df_filtro_m[columna_texto].apply(filtrar_registros)]
```

#Preprocesamiento

```python
df_filtro_mx = pd.read_csv('/content/drive/MyDrive/GDELT_API3/Datos/noticias_sin_duplicados.csv')
df_filtro_mx.info()
```

```python
import itertools
import re

df_filtro_mx['texto_completo']=df_filtro_mx['texto_completo'].apply(str)

def clean_text(texto):
  texto = re.sub(r'<[^>]*?>', '', texto) # Eliminar caracteres HTML
  #texto = re.sub(r'[0-9]+', '', texto) #eliminar números
  texto = re.sub(r"[^\w\s]", "", texto) #eliminar puntuación
  texto = re.sub(r'^https?:\/\/.*[\r\n]*', '', texto, flags=re.MULTILINE) #eliminar URLS
  texto = re.sub("_", "", texto)
  texto = re.sub(r'\+', '', texto)
  texto = re.sub(r'\:', '', texto)

  return texto
```

```python
df_filtro_mx['texto_completo'] = df_filtro_mx['texto_completo'].apply(clean_text)
df_filtro_mx['texto_completo']
```
